```python
import pandas as pd
import numpy as np
import random
```

```python
dates = pd.date_range('2021-12-01 7:00', '2022-01-01 00:00')
dates
```

```python
# Create a daily 7am - 12am timestamp every 5 mins
ls = []
for i in dates:
    daily_business_hour = pd.date_range(i, i+pd.Timedelta('17 hours'), freq='5min')
    [ls.append(i) for i in daily_business_hour]
log = pd.DataFrame(ls)
log
```

```python
log.columns = ["timestamp"]
log.head(3)
```

```python
log["wm_id"] = 1
log.head(3)
```

```python
log["in_use"] = False
log.head(3)
```

```python
from random import choices
status = [0, 1]
weights = [0.9, 0.1]
choices(status, weights)
```

```python
# Generate a random boolean column. When a row=True, it lasts for 50 mins (other rows within the 50 mins will be True)
log["in_use"] = False
i=0
while i < len(log):
    random_bool = bool(choices(status, weights)[0])
    # print(i, '\t', random_bool)
    if random_bool:
        # print('i when True=', i)
        wm_ends = log['timestamp'][i] + pd.Timedelta('50 min')
        while i < len(log) and log['timestamp'][i] <= wm_ends:
            log['in_use'][i] = random_bool
            i+=1   
        # print('i after LOOP=', i, '\t in if block.') 
    i+=1    
    # print('i after True=', i, '\t outside of loop.') 
    
log
```
